#### 1. Without the use of any python libraries or modules, develop a function that can perform matrix addition, given two numpy matrices.

```python
def matrix_addition(matrix1, matrix2):
    # Check if the matrices have the same dimensions
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        raise ValueError("Matrices must have the same dimensions for addition.")
    
    # Get the dimensions of the matrices
    rows = len(matrix1)
    cols = len(matrix1[0])
    
    # Initialize a result matrix with zeros
    result_matrix = [[0 for _ in range(cols)] for _ in range(rows)]
    
    # Perform matrix addition element-wise
    for i in range(rows):
        for j in range(cols):
            result_matrix[i][j] = matrix1[i][j] + matrix2[i][j]
    
    return result_matrix

# Example usage:
matrix_A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix_B = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]

result = matrix_addition(matrix_A, matrix_B)
for row in result:
    print(row)
```

#### 2. Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.

```python
def matrix_multiplication(matrix1, matrix2):
    # Check if the matrices can be multiplied (number of columns in matrix1 equals number of rows in matrix2)
    if len(matrix1[0]) != len(matrix2):
        raise ValueError("Matrices cannot be multiplied. Number of columns in matrix1 must equal number of rows in matrix2.")
    
    # Get the dimensions of the matrices
    rows1, cols1 = len(matrix1), len(matrix1[0])
    rows2, cols2 = len(matrix2), len(matrix2[0])
    
    # Initialize a result matrix with zeros
    result_matrix = [[0 for _ in range(cols2)] for _ in range(rows1)]
    
    # Perform matrix multiplication
    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]
    
    return result_matrix

# Example usage:
matrix_A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix_B = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]

result = matrix_multiplication(matrix_A, matrix_B)
for row in result:
    print(row)
```

#### 3. The rule of distributivity states that given two matrices  ùê¥ and  ùêµ and a scalar,  ùëò, then  ùëò(ùê¥+ùêµ)=ùëòùê¥+ùëòùêµ. Instead of writing a proof mathematically, develop two codes for  ùëò(ùê¥+ùêµ) and  ùëòùê¥+ùëòùêµ.

```python
def distributive_property_1(k, matrix_A, matrix_B):
    # Check if the matrices have the same dimensions
    if len(matrix_A) != len(matrix_B) or len(matrix_A[0]) != len(matrix_B[0]):
        raise ValueError("Matrices must have the same dimensions for addition.")
    
    # Get the dimensions of the matrices
    rows, cols = len(matrix_A), len(matrix_A[0])
    
    # Initialize result matrix with zeros
    result_matrix = [[0 for _ in range(cols)] for _ in range(rows)]
    
    # Perform matrix addition element-wise
    for i in range(rows):
        for j in range(cols):
            result_matrix[i][j] = k * (matrix_A[i][j] + matrix_B[i][j])
    
    return result_matrix

# Example usage:
k_value = 2
matrix_A = [[1, 2], [3, 4]]
matrix_B = [[5, 6], [7, 8]]

result_1 = distributive_property_1(k_value, matrix_A, matrix_B)
for row in result_1:
    print(row)


print("\n\n")  
def distributive_property_2(k, matrix_A, matrix_B):
    # Check if the matrices have the same dimensions
    if len(matrix_A) != len(matrix_B) or len(matrix_A[0]) != len(matrix_B[0]):
        raise ValueError("Matrices must have the same dimensions for addition.")
    
    # Get the dimensions of the matrices
    rows, cols = len(matrix_A), len(matrix_A[0])
    
    # Initialize result matrices with zeros
    result_matrix_A = [[0 for _ in range(cols)] for _ in range(rows)]
    result_matrix_B = [[0 for _ in range(cols)] for _ in range(rows)]
    
    # Perform matrix multiplication with scalar element-wise
    for i in range(rows):
        for j in range(cols):
            result_matrix_A[i][j] = k * matrix_A[i][j]
            result_matrix_B[i][j] = k * matrix_B[i][j]
    
    # Perform matrix addition element-wise
    result_matrix = [[result_matrix_A[i][j] + result_matrix_B[i][j] for j in range(cols)] for i in range(rows)]
    
    return result_matrix

# Example usage:
k_value = 2
matrix_A = [[1, 2], [3, 4]]
matrix_B = [[5, 6], [7, 8]]

result_2 = distributive_property_2(k_value, matrix_A, matrix_B)
for row in result_2:
    print(row)

```

#### 4. Without using a python library or modules develop a function that can extract the diagonal of a numpy matrix.

```python
def extract_diagonal(matrix):
    # Check if the matrix is square
    if len(matrix) != len(matrix[0]):
        raise ValueError("Matrix must be square to extract the diagonal.")
    
    # Get the length of the matrix
    n = len(matrix)
    
    # Extract the diagonal elements
    diagonal = [matrix[i][i] for i in range(n)]
    
    return diagonal

# Example usage:
matrix_A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

diagonal_elements = extract_diagonal(matrix_A)
print(diagonal_elements)
```

#### 5. Without using a python library or modules develop a function that can find a trace of a numpy matrix.

```python
def matrix_trace(matrix):
    # Check if the matrix is square
    if len(matrix) != len(matrix[0]):
        raise ValueError("Matrix must be square to find the trace.")
    
    # Get the length of the matrix
    n = len(matrix)
    
    # Calculate the trace (sum of diagonal elements)
    trace_sum = sum(matrix[i][i] for i in range(n))
    
    return trace_sum

# Example usage:
matrix_A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

trace_value = matrix_trace(matrix_A)
print(trace_value)
```
