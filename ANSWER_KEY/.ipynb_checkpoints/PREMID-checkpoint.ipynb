1. Show that when $\cos{(\theta)}=0$, the vectors are orthogonal.

Solution:
  
  By definition, two vectors are orthogonal $\iff \textbf{u} \cdotp \textbf{v} = 0$.

  And by definition of a cosine of a dot product:
  \begin{equation}
  \cos{(\theta)} = \frac{\textbf{u} \cdotp \textbf{v}}{||u| ||v||}
  \end{equation}

  since $||u|$ and $||v||$ cannot always be a zero vector:
  \begin{equation}
  \cos{(\theta)} = \frac{\textbf{u} \cdotp \textbf{v}}{||u| ||v||}
  \end{equation} 
  can only be zero $\iff$the vectors $u$ and $v$ are orthogonal, $ \textbf{u} \cdotp \textbf{v} = 0$

3. Show that when $ad-bc$ is not zero, the systems of linear equations $ax+by=r$ and $cx+dy=s$ has a unique solution.
Solution:

The hint here is "$ad-bc$ is not zero".
\begin{equation}
\begin{bmatrix}
\begin{array}{cc|c}
a & b & r \\
c & d & s \\
\end{array}
\end{bmatrix}
\xrightarrow[\frac{R_1}{a}]{\frac{R_2}{c}}
\begin{bmatrix}
\begin{array}{cc|c}
1 & \frac{b}{a} & \frac{r}{a} \\
1 & \frac{d}{c} & \frac{s}{b} \\
\end{array}
\end{bmatrix}
\xrightarrow{R_2-R_1}
\begin{bmatrix}
\begin{array}{cc|c}
1 & \frac{b}{a} & \frac{r}{a} \\
0 & \frac{d}{c}-\frac{b}{a} & \frac{s}{b}-\frac{r}{a} \\
\end{array}
\end{bmatrix}
\xrightarrow{simplify}
\begin{bmatrix}
\begin{array}{cc|c}
1 & \frac{b}{a} & \frac{r}{a} \\
0 & \frac{ad-bc}{ac} & \frac{as-br}{ab} \\
\end{array}
\end{bmatrix}
\end{equation}

- **NOTE:** As you can observe the term $ad-bc$ sprang up, and by condition it is not equals to zero, hence the equation is not redundant. We let $\alpha = \frac{ad-bc}{ac}$ and $\beta = \frac{as-br}{ab}$, $\gamma = \frac{b}{a}$, and $\theta = \frac{r}{a}$:
\begin{equation}
\begin{bmatrix}
\begin{array}{cc|c}
1 & \gamma & \theta \\
0 & \alpha & \beta \\
\end{array}
\end{bmatrix}
\xrightarrow{\frac{R_2}{\alpha}}
\begin{bmatrix}
\begin{array}{cc|c}
1 & \gamma & \theta \\
0 & 1 & \frac{\beta}{\alpha} \\
\end{array}
\end{bmatrix}
\xrightarrow{R_1 - \gamma R_2}
\begin{bmatrix}
\begin{array}{cc|c}
1 & 0 & \theta - \frac{\beta \gamma}{\alpha} \\
0 & 1 & \frac{\beta \gamma}{\alpha} \\
\end{array}
\end{bmatrix}
\end{equation}
- Hence:
\begin{equation}
\begin{bmatrix}
\begin{array}{c|c}
x & \theta - \frac{\beta \gamma}{\alpha} \\
y & \frac{\beta \gamma}{\alpha} \\
\end{array}
\end{bmatrix}
\end{equation}
- $\therefore$ the system of linear equation has a unique solution.
